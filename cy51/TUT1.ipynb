```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob"],
    "Math": [85, 40],
    "AI": [90, 55],
    "ML": [88, 35]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Compute average marks
df["Average"] = df[["Math", "AI", "ML"]].mean(axis=1)

# Add Result column
df["Result"] = df["Average"].apply(lambda x: "Pass" if x >= 50 else "Fail")

print("\nWith Averages & Results:\n", df)
```

```python
import pandas as pd

timeline = {
    "Year": [1956, 1997, 2011, 2016, 2023],
    "Milestone": [
        "Dartmouth Conference (Birth of AI)",
        "IBM Deep Blue beats Garry Kasparov",
        "IBM Watson wins Jeopardy!",
        "AlphaGo defeats Lee Sedol",
        "ChatGPT gains global adoption"
    ]
}

df_timeline = pd.DataFrame(timeline)
df_timeline = df_timeline.sort_values("Year")
print("AI History Timeline:\n", df_timeline)
```
