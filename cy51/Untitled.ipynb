```python
import pandas as pd
import numpy as np


data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "Roll_No": [101, 102, 103, 104, 105, 106],
    "Marks": [85, 78, 92, 74, 88, 95],
    "Age": [23, 29, 35, 42, 28, 33],
    "Department": ["IT", "HR", "Finance", "IT", "HR", "Finance"],
    "Salary": [70000, 50000, 65000, 72000, 48000, 80000],
    "Joining_Date": pd.to_datetime([
        "2019-05-21", "2021-03-15", "2020-07-19",
        "2022-01-10", "2018-11-05", "2023-04-12"
    ])
}
df = pd.DataFrame(data)
print("1:\n", df)
```

```python
print("\n4 Shape:", df.shape)
print("4 Columns:", df.columns.tolist())
print("4 Data Types:\n", df.dtypes)

print("\n5 Summary Stats:\n", df.describe(include="all"))
```

```python
print("\n6:\n", df["Name"])

print("\n7:\n", df[["Name", "Salary"]])

print("\n8 head():\n", df.head(3))
print("8 iloc:\n", df.iloc[:3])

print("\n9:\n", df[df["Age"] > 30])

print("\n10:\n", df[df["Department"] == "HR"])
```

```python
print("\n11:\n", df.sort_values(by="Salary", ascending=False))

print("\n12:\n", df.sort_values(by=["Department", "Age"]))

print("\n13:\n", df[(df["Salary"] > 60000) & (df["Department"] == "IT")])

print("\n14:\n", df[df["Joining_Date"] > "2020-01-01"])

print("\n15:\n", df[df["Name"].str.startswith("A")])
```

```python

print("\n16:\n", df.groupby("Department")["Salary"].mean())

print("\n17:\n", df["Department"].value_counts())

print("\n18:\n", df.groupby("Department")["Salary"].agg(["min", "max"]))

print("\n19:\n", df.groupby("Department")["Age"].mean())

df["Joining_Year"] = df["Joining_Date"].dt.year
print("\n20:\n", df.groupby(["Department", "Joining_Year"]).size())
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Employee dataset (reused from earlier)
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "Age": [23, 29, 35, 42, 28, 33],
    "Department": ["IT", "HR", "Finance", "IT", "HR", "Finance"],
    "Salary": [70000, 50000, 65000, 72000, 48000, 80000]
}
df = pd.DataFrame(data)
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.figure()
plt.plot(x, y)
plt.title("1: Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python

x = np.arange(0, 11)
plt.figure()
plt.plot(x, x**2)
plt.title("2: y = x^2")
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
plt.figure()
plt.plot(x, np.sin(x), label="Sine")
plt.plot(x, np.cos(x), label="Cosine")
plt.title("3: Sine & Cosine")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(x, np.sin(x), 'r--', label="Sine")
plt.plot(x, np.cos(x), 'g-', label="Cosine")
plt.title("4: With Legends & Grid")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 11)
fig, axs = plt.subplots(1, 2, figsize=(10,4))
axs[0].plot(x, x**2)
axs[0].set_title("5: y = x^2")
axs[1].plot(x, x**3)
axs[1].set_title("5: y = x^3")
plt.show()
```

```python

subjects = ["Math", "Science", "English", "History", "CS"]
marks = [85, 78, 92, 74, 88]
plt.figure()
plt.bar(subjects, marks)
plt.title("6: Student Marks")
plt.show()

avg_salary = df.groupby("Department")["Salary"].mean()
plt.figure()
avg_salary.plot(kind="bar")
plt.title("7: Avg Salary by Dept")
plt.show()
```

```python
dept_counts = df["Department"].value_counts()
plt.figure()
dept_counts.plot(kind="barh")
plt.title("8: Employee Count by Dept")
plt.show()

plt.figure()
dept_counts.plot(kind="pie", autopct='%1.1f%%')
plt.title("9: Dept Distribution")
plt.ylabel("")
plt.show()
```

```python

students = ["A", "B", "C", "D", "E"]
marks1 = [80, 70, 90, 85, 60]
marks2 = [75, 65, 95, 80, 70]
plt.figure()
plt.bar(students, marks1, label="Test1")
plt.bar(students, marks2, bottom=marks1, label="Test2")
plt.title("10: Stacked Marks")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.hist(df["Age"], bins=5, color="skyblue", edgecolor="black")
plt.title("11: Age Histogram")
plt.show()

plt.figure()
plt.hist(df["Salary"], bins=20, color="orange", edgecolor="black")
plt.title("12: Salary Histogram (20 bins)")
plt.show()
```

```python
plt.figure()
n, bins, patches = plt.hist(df["Age"], bins=5, color="skyblue", edgecolor="black")

# Make the middle bar bigger (change its facecolor)
mid_index = len(patches) // 2
patches[mid_index].set_facecolor("orange")
patches[mid_index].set_edgecolor("red")
patches[mid_index].set_linewidth(2)

plt.title("11: Age Histogram (Middle Bar Highlighted)")
plt.show()
```
